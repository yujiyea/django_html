```python
import re
import requests
from bs4 import BeautifulSoup
url = 'https://search.naver.com/search.naver?sm=top_hty&fbm=1&ie=utf8&query=sdf'
web = requests.get(url)
soup = BeautifulSoup(web.content,"html.parser")
print(soup)
```

```python
forms = soup.find_all('form')
form_name =[]
```

```python
for form in forms:
    links = form.find_all('input', attrs={'type':"text"}) #검색
    print(links)
    for link in links:
        form_name = link['name'] #찾은 입력창의 이름
```
